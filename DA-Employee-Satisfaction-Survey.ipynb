### Finding Outliers

```python
sns.boxplot(x=df['Age'])
plt.show()
```

```python
# Age Variable has no Outliers
```

```python
sns.boxplot(x=df['Experience'])
plt.show()
```

```python
sns.boxplot(x=df['PhysicalActivityHours'])
plt.show()
```

* PhysicalActivityHours Variable has Outliers

```python
#To extract the Outliers--

Q1= df['PhysicalActivityHours'].quantile(0.25)
print('Q1: ', Q1)

Q3= df['PhysicalActivityHours'].quantile(0.75)
print('Q3: ', Q3)

IQR= Q3-Q1
print('IQR: ', IQR)

#lower_limit

ll= Q1-(IQR*1.5)
print('lower_limit: ', ll)

#upper_limit

ul= Q3+(IQR*1.5)
print('upper_limit: ', ul)
```

```python
df2=df[(df['PhysicalActivityHours']< ll) | (df['PhysicalActivityHours']> ul)]
pd.set_option('display.max_columns', None)
df2
```

* The no. of Outliers is low and they do deviate much from the upper limit. Hence, we will Replace them.

```python
df['PhysicalActivityHours']= df['PhysicalActivityHours'].clip(upper=4.8)
df.head()
```

### Insights-

Overall-

* Employees with highest Stress, workload and lowest WLB, Sleephours have lowest Job Satisfaction
* Employees with PhD have lower Job Satisfaction.
* Males have lower Job Satisfaction
* Employees doing Overtime have lower Job Satisfaction.
* Employees having the lowest Work Environment Quality have the lowest Job Satisfaction.

Department Wise- 

IT-Dept-

* IT Dept has the lowest Job Satisfaction.
* IT Dept employees do the most amount of Over Time work and have the lowest Sleep Hours.
* Average Team size is 16 in a mid-sized IT company which might obstruct promotion and growth of employees. The ideal avg. team size should be 5-7 employees.
* Most Senior level employees have low Job Satisfaction. Especially Lead level employees have the lowest Job Satisfaction
* Full-time employees have lower Job Satisfaction compared to part-time or contract based employees.


Other Departments-

* Customer Service and HR employees have the lowest WLB
* Marketing employees have the lowest Work Environment rating.

### Recommendations-

* Provide WFH option to employees having Commute Distance above 8-10 kms.
* Keep the avg team size to 5-7 employees.
* Provide increment to PhD employees.
* Provide attractive incentives to employees doing Overtime.
* Improve the work environment quality by providing amenities to employees.
* Provide Mental health services for employees to manage stress, workload and improve WLB, Sleephours.
